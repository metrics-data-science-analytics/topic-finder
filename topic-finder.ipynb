```python
# This is for installing the library I used to scrape tweets
# !pip install git+https://github.com/JustAnotherArchivist/snscrape.git
# !pip install twitter_scraper

import os
import pandas as pd
from twitter_scraper import get_tweets
```

```python
tweet_count = 1000;
text_query = "covid"
since_date = "2021-02-21"
until_date = "2021-03-20"

os.system('snscrape --jsonl --max-results {} --since {} twitter-search "{} until:{}"> covid-query-one.json'.format(tweet_count, since_date, text_query, until_date))
```

```python
tweets_set_1 = pd.read_json('covid-query-one.json', lines=True)

tweets_set_1.describe()
```
